```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
species = ['H_meleagridis', 'T_foetus', 'T_vaginalis']
```

```python
names = ["Number of genes per-species in orthogroup"] + species
```

```python
df = pd.read_csv('../resource/Proteome/OrthoFinder/Results_Feb22/Comparative_Genomics_Statistics/Statistics_PerSpecies.tsv', sep='\t', skiprows=58, nrows=20, names=names, header=None)
```

```python
df.head(50)
```

```python
df.set_index('Number of genes per-species in orthogroup').plot(kind='barh', stacked=True, figsize=(10, 7))

plt.title('Orthogroup Statistics')
plt.ylabel('Number of Genes per Species in Orthogroup')
plt.xlabel('Genes Assigned to Orthogroups')
plt.xticks(rotation=45)
plt.legend(title='Species')
plt.tight_layout()

plt.savefig("OrthoGroupStatistics.png", transparent=True)
```
