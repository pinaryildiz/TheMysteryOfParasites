```python
import matplotlib.pyplot as plt
```

```python
plt.style.use("default")
```

```python
# RepeatMasker sonuçlarındaki tekrar türlerinin uzunlukları
repeat_lengths = {
    'Sınıflandırılmamış': 4830877,  # 7.15% of sequence
    'Basit Tekrarlar': 1222891,  # 1.81% of sequence
}

# Tekrar türlerinin isimleri ve uzunlukları
labels = repeat_lengths.keys()
sizes = repeat_lengths.values()

# Pasta grafiği oluşturma
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.2f%%', startangle=90)

# Çember yerine biraz daha oval bir pasta grafiği elde etmek için 'equal' aspect ratio'sunu ayarlayın
ax1.axis('equal')

plt.title('RepeatMasker Sonuçlarına Göre Maskelenme Dağılımı')
plt.savefig(f"foetus_d.png", transparent=True)
```

```python
total = 67585742
masked = 6053768
unmasked = total - masked
repeat_lengths = {
    'Maskelenmemiş': unmasked,
    'Maskelenmiş': masked,
}

labels = repeat_lengths.keys()
sizes = repeat_lengths.values()

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.2f%%', startangle=90)

ax1.axis('equal')

plt.title('RepeatMasker Sonuçlarına Göre Maskelenme Yüzdesi')
plt.savefig(f"foetus_t.png", transparent=True)
```
